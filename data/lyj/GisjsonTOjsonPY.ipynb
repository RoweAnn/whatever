```python
#GisjsonTOjson
#2024年4月1日 yy 但是 中文字符被转为其他东西了 
import json

# Open the first JSON file and load data
with open('XM_XA建筑_FeaturesToJSON.json', 'r', encoding='utf-8') as f:
    data = json.load(f)

# Extract NAME, lon, and lat from each feature
locations = {}
for feature in data['features']:
    name = feature['attributes']['NAME']
    lon = feature['attributes']['lon']
    lat = feature['attributes']['lat']
    locations[name] = f"{lat},{lon}"

# Write the extracted data into the structure of the second JSON file
with open('geocoders.json', 'w') as f:
    json.dump(locations, f, indent=2)
```

# 输出建筑轮廓文件

```python
#yy 01  输出XMXA_modified.geojson
import json

# 截取小数部分函数
def truncate_float(f, n):
    s = '{}'.format(f)
    if 'e' in s or 'E' in s:
        return '{0:.{1}f}'.format(f, n)
    i, p, d = s.partition('.')
    return '.'.join([i, (d+'0'*n)[:n]])

# 读取geojson文件
with open('XMXA.geojson', 'r', encoding='utf-8') as f:
    geojson_data = json.load(f)

# 遍历features
for feature in geojson_data['features']:
    # 去除坐标数组中每个点的第三个值
    for coordinates in feature['geometry']['coordinates']:
        for coord in coordinates:
            # 截取小数部分
            coord[0] = truncate_float(coord[0], 11)
            coord[1] = truncate_float(coord[1], 14)
            coord.pop(2)
            
    # 添加"color"字段
    feature['properties']['color'] = "#255294"
    # 替换"HEIGHT"为"max_height"
    feature['properties']['max_height'] = feature['properties'].pop('HEIGHT')

    # 移除不需要的字段
    properties_to_remove = ["OBJECTID", "NAME", "SHAPE_AREA", "SHAPE_LEN", "lon", "lat"]
    for prop in properties_to_remove:
        feature['properties'].pop(prop, None)

# 写入修改后的geojson数据到新文件，去除双引号并保持单行形式
with open('trXMXA_modified.geojson', 'w', encoding='utf-8') as f:
    json.dump(geojson_data, f, ensure_ascii=False, separators=(',', ':'))
```

```python
#yy 02 去除坐标中经纬度值的双引号和方括号 去除max_height值的双引号 
#XMXA_modified_fixed.geojson 中间文件
import json

# 读取geojson文件
with open('trXMXA_modified.geojson', 'r') as f:
    data = json.load(f)

# 遍历features
for feature in data['features']:
    # 去除坐标中经纬度值的双引号和方括号
    coordinates = feature['geometry']['coordinates'][0]
    for i in range(len(coordinates)):
        coordinates[i][0] = float(coordinates[i][0].strip('"[]')) - 1.499  # 减去1.49
        coordinates[i][1] = float(coordinates[i][1].strip('"[]'))

    # 去除max_height值的双引号
    feature['properties']['max_height'] = float(feature['properties']['max_height'].strip('"'))

# 保存修改后的数据到新文件
with open('XMXA_modified_fixed.geojson', 'w') as f:
    json.dump(data, f)
```

```python
import json

# 读取原始JSON文件
with open('geocoderslyj0.json', 'r', encoding='utf-8') as f:
    data = json.load(f)

# 修改坐标中的精度
for key, value in data.items():
    lat, lon = value.split(',')
    lon = float(lon) - 1.499
    data[key] = f"{lat},{lon}"

# 将修改后的数据写入新的JSON文件
with open('geocoderslyj.json', 'w', encoding='utf-8') as f:
    json.dump(data, f, ensure_ascii=False, indent=2)
```
